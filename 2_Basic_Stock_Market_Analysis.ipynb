```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import yfinance as yf
```

1) Download historical stock data for Apple, Microsoft, Amazon, and Google

```python
tickers = ['AAPL', 'MSFT', 'AMZN', 'GOOG']
stock_data = yf.download(tickers, start='2015-01-01', end='2020-12-31')
```

2) Analayze & Plot the Data

```python
stock_data['Adj Close'].plot()
plt.show()
```

3) Calculate & Plot Daily Returns for Apple

```python
daily_returns = stock_data['Adj Close'].pct_change()

daily_returns.hist(bins=100)
plt.show()
```

```python
daily_returns['AAPL'].hist(bins=100)
plt.ylabel('Frequency')
plt.xlabel('Daily Return')
plt.title('AAPL')
plt.show()
```

4) Calculate & Plot Some Performance Metrics

```python
mean_returns = daily_returns.mean()
std_returns = daily_returns.std()

plt.scatter(std_returns, mean_returns)
for label,x,y in zip(daily_returns.columns, std_returns, mean_returns):
    plt.annotate(label,xy=(x,y),xytext=(10,0),textcoords='offset points')
plt.title('Mean vs. Standard Deviation')
plt.ylabel('Mean Daily Return')
plt.xlabel('Standard Deviation')
plt.show()
```
